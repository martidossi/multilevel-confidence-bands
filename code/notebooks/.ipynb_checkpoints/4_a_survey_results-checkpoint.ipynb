### 1.2 Time spent on task

```python
# Filter time-related columns
df_time = df.copy()

for house_name in df_houses.name:
    house_name = house_name.lower()
    df_time[f'{house_name}_duration'] = (
        df_time[f'{house_name}_last_click'] - df_time[f'{house_name}_first_click']
    )

df_time = df_time[
    id_cols
    + [f'{house_name.lower()}_duration' for house_name in df_houses.name]
    + [f'{house_name.lower()}_click_count' for house_name in df_houses.name]
    + [f'{house_name.lower()}_first_click' for house_name in df_houses.name]
    + [f'{house_name.lower()}_page_submit' for house_name in df_houses.name]
]

df_time.head()
```

```python
# Define the metrics we want to extract
metrics = ["click_count", "duration", "first_click", "page_submit"]

collectors = {m: [] for m in metrics}
for house in df_houses.name:
    house_lower = house.lower()
    house_complexity = df_houses.loc[df_houses.name == house, "complexity"].values[0]

    for metric in metrics:
        tmp = df_time.copy()
        old_col = f"{house_lower}_{metric}"
        tmp = tmp[id_cols + [old_col]].rename(columns={old_col: metric})
        tmp["house_name"] = house_lower
        tmp["complexity"] = house_complexity
        collectors[metric].append(tmp)

# final dataframes
df_click_count = pd.concat(collectors["click_count"])
df_duration = pd.concat(collectors["duration"])
df_first_click = pd.concat(collectors["first_click"])
df_page_submit = pd.concat(collectors["page_submit"])
```

**Additional analysis on time**

Relationship between accuracy and estimation time.

```python
df_house_properties = df_houses.copy()
df_house_properties['name'] = df_house_properties['name'].str.lower()

df_house_melted = (
    df[['participant_id'] + [col for col in df.columns if 'house_value' in col]]
        .drop(['training_house_value'], axis=1)
        .melt(
            id_vars='participant_id',
            value_vars=[col for col in df.columns if 'house_value' in col if col!="training_house_value"],
            var_name='house_name',
            value_name='house_value'
        )
)
df_house_melted['house_name'] = df_house_melted['house_name'].str.replace('_house_value', '', regex=False)
df_house_melted['house_value'] = df_house_melted['house_value']/100
df_house_melted = df_house_melted.rename(columns={'house_value': 'house_pred'})
df_house_melted = df_house_melted.merge(df_house_properties[['name', 'true_value']], left_on='house_name', right_on='name', how='left').drop(['name'], axis=1)

df_house_melted['residual_error'] = df_house_melted["true_value"] - df_house_melted["house_pred"]
df_house_melted['abs_error'] = np.abs(df_house_melted['residual_error'])
df_house_melted[df_house_melted.participant_id=='participant_1']
```

```python
df_time_aug = df_page_submit.copy()
df_time_aug = df_time_aug.merge(df_house_melted, on=['house_name', 'participant_id'], how='outer')
df_time_aug = df_time_aug.rename(columns={'page_submit': 'response_time'})
df_time_aug[df_time_aug.participant_id=='participant_1']
```

```python
time_results = []
for _cond in df_time_aug['condition'].unique():
    for _comp in df_time_aug['complexity'].unique():
        _subset = df_time_aug[(df_time_aug['condition'] == _cond) & (df_time_aug['complexity'] == _comp)]
        _rho, _p = spearmanr(_subset['response_time'], _subset['residual_error'])
        time_results.append({'condition': _cond, 'complexity': _comp, 'rho': _rho, 'p': _p})

time_corr = pd.DataFrame(time_results)
print(time_corr.round(3))
```

## 2. Estimation results by condition (per house)

```python
error_bars = list(df[df.condition=='Error bars'][f'{sel_house.lower()}_house_value'])
confidence_bands = list(df[df.condition=='Confidence bands'][f'{sel_house.lower()}_house_value'])
no_ai = list(df[df.condition=='No AI'][f'{sel_house.lower()}_house_value'])
```

```python
def find_closest(data, true_val):
    return min(data, key=lambda x: abs(x - true_val))

closest_eb = find_closest(error_bars, true_value)
closest_cb = find_closest(confidence_bands, true_value)
closest_noai = find_closest(no_ai, true_value)

bins = np.arange(0, 810, 50)
# counts1, _ = np.histogram(error_bars, bins=bins)
```

```python
df_counts = pd.DataFrame(data=list(zip(bins, [0]*len(bins))), columns=['val', 'count_tmp']).set_index('val')
df_counts_error_bars = pd.Series(error_bars).value_counts().reset_index()
df_counts_confidence_bands = pd.Series(confidence_bands).value_counts().reset_index()
df_counts_no_ai = pd.Series(no_ai).value_counts().reset_index()

counts1 = list(df_counts.join(df_counts_error_bars.set_index('index')).fillna(0)['count'])
counts2 = list(df_counts.join(df_counts_confidence_bands.set_index('index')).fillna(0)['count'])
counts3 = list(df_counts.join(df_counts_no_ai.set_index('index')).fillna(0)['count'])
```

```python
plt.figure(figsize=(12, 6))

bar_width = 18
group_width = 50
group_spacing = 30

# Adjust the x-positions for better grouping
x_vals = np.arange(0, len(bins) * (group_width + group_spacing), group_width + group_spacing)

x_eb = x_vals - group_width / 3
x_cb = x_vals
x_noai = x_vals + group_width / 3

# Alternated background
for i, val in enumerate(x_vals):
    if i % 2 == 0:
        plt.axvspan(
            val - group_width / 2 - group_spacing / 2,
            val + group_width / 2 + group_spacing / 2,
            color="lightgrey", alpha=0.2
        )

eb_bars = plt.bar(
    x_eb, counts1, width=bar_width, label="Error bars",
    color=condition_color_palette["Error bars"],
    edgecolor="black", linewidth=0.8
)
cb_bars = plt.bar(
    x_cb, counts2, width=bar_width, label="Confidence bands",
    color=condition_color_palette["Confidence bands"],
    edgecolor="black", linewidth=0.8
)
noai_bars = plt.bar(
    x_noai, counts3, width=bar_width, label="No AI",
    color=condition_color_palette["No AI"],
    edgecolor="black", linewidth=0.8
)

def mark_best(bars, closest_val, color, bins, x_shift):
    closest_bin = np.digitize([closest_val], bins, right=True)[0]
    if 0 <= closest_bin < len(bars):
        bar = bars[closest_bin]
        bar_x = bar.get_x() + bar.get_width() / 2
        bar_height = bar.get_height()
        plt.text(bar_x, bar_height + 0.3, "x",
                 ha="center", va="bottom",
                 fontsize=16, color=color)

mark_best(eb_bars, closest_eb, "black", bins, -bar_width)
mark_best(cb_bars, closest_cb, "black", bins, 0)
mark_best(noai_bars, closest_noai, "black", bins, bar_width)

# Adjust the x-axis labels
plt.xticks(
    ticks=x_vals,
    labels=[f"{int(v)}k" for v in np.arange(0, 801, 50)]
)

plt.yticks(np.arange(0, 21, 2))
plt.ylim(0, 22)

plt.xlabel("Estimated value ($)", fontsize=12)
plt.ylabel("Number of guesses", fontsize=12)
plt.title(
    f"{sel_house.capitalize()}\n"
    f"Estimation distribution (true value: {true_value}k$)",
    fontsize=14
)

plt.grid(True, axis="y", linestyle="--", alpha=0.5)

plt.legend(
    ncol=1,
    loc="upper right",
    fontsize=11,
    frameon=True
)

plt.tight_layout()
plt.savefig(results_folder_path + f"estimation_results/estimation_distribution_{sel_house.lower()}.pdf")
plt.show()
```

### 2.1 Error distribution by condition (per house, complexity level, all)

```python
# By house
df["sel_house_residual_error"] = df[f"{sel_house.lower()}_house_value"] - true_value

fig_err_dist, ax_err_dist = plt.subplots(1, 1, figsize=(7, 5))

for i, cond in enumerate(boxplots_order):
    data = df.loc[df["condition"] == cond, "sel_house_residual_error"].values

    if len(data) > 1:  # gaussian_kde needs at least 2 data points
        kde = gaussian_kde(data)
        x_vals = np.linspace(min(data), max(data), 200)
        y_vals = kde(x_vals)

        # Scale KDE to fit nicely next to boxplot
        y_vals = y_vals / y_vals.max() * 0.3

        # Shift KDE to the right of the boxplot
        ax_err_dist.plot(i + y_vals, x_vals, color="black", alpha=0.7,
                         linestyle=":", linewidth=1.5, zorder=3)


# Boxplot
sns.boxplot(
    x="condition", y="sel_house_residual_error", data=df,
    order=boxplots_order,
    showfliers=False, hue="condition", palette=condition_color_palette, width=0.5,
    boxprops=dict(edgecolor="black", linewidth=0.8),
    whiskerprops=dict(color="black", linewidth=0.8),
    capprops=dict(color="black", linewidth=0.8),
    medianprops=dict(color="black", linewidth=1),
)

sns.stripplot(
    x="condition", y="sel_house_residual_error", data=df,
    jitter=True, size=5,
    edgecolor="black",
    palette=condition_color_palette,
    hue="condition",
    linewidth=0.5,
    alpha=0.6
)

ax_err_dist.axhline(0, color="black", linestyle="--", linewidth=1.5, alpha=0.8)
ax_err_dist.set_xlabel("")
ax_err_dist.set_ylabel("Residual error ($)")

ax_err_dist.yaxis.set_major_formatter(
    mtick.FuncFormatter(lambda x, pos: f"{int(x):,}k")
)

ax_err_dist.set_xticks(ax_err_dist.get_xticks())
ax_err_dist.set_xticklabels(
    [mapping_labels[l.get_text()] for l in ax_err_dist.get_xticklabels()]
)

plt.title(f"{sel_house.capitalize()}\nError distribution", fontsize=10)
plt.grid(True, axis="y", linestyle="--", alpha=0.5)

plt.savefig(results_folder_path + f"error_distribution/error_distribution_{sel_house.lower()}.pdf")
plt.show()
```

```python
# All together (for the paper)

fig_boxplots, axes_boxplots = plt.subplots(
    1, len(complexity_levels),
    figsize=(14, 4),
    sharey=True
)

for idx, complexity in enumerate(complexity_levels):
    houses = list(df_houses[df_houses.complexity == complexity].name)

    df_class = df.copy()
    true_vals = (
        df_houses[df_houses.complexity == complexity]
        .set_index("name")["true_value"]
        .to_dict()
    )

    df_class = df_class[["condition"] + [f"{house.lower()}_house_value" for house in houses]]
    for house in houses:
        df_class[f"{house.lower()}_house_value"] -= true_vals[house] * 100

    df_class = pd.melt(
        df_class,
        id_vars="condition",
        value_vars=[f"{house.lower()}_house_value" for house in houses]
    ).rename(columns={"value": "residual_error"})

    ax_boxplots = axes_boxplots[idx]

    # KDE overlays
    for i, cond in enumerate(boxplots_order):
        data = df_class.loc[df_class["condition"] == cond, "residual_error"].values
        if len(data) > 1:
            kde = gaussian_kde(data)
            x_vals = np.linspace(min(data), max(data), 200)
            y_vals = kde(x_vals)
            y_vals = y_vals / y_vals.max() * 0.3
            ax_boxplots.plot(
                i + y_vals, x_vals,
                color="black", alpha=0.7,
                linestyle=":", linewidth=1.5, zorder=3
            )

    # Boxplot
    sns.boxplot(
        x="condition", y="residual_error", data=df_class,
        order=boxplots_order,
        showfliers=False, hue="condition",
        palette=condition_color_palette, width=0.5,
        boxprops=dict(edgecolor="black"),
        whiskerprops=dict(color="black"),
        capprops=dict(color="black"),
        medianprops=dict(color="black", linewidth=1),
        ax=ax_boxplots, legend=False
    )

    # Stripplot overlay
    sns.stripplot(
        x="condition", y="residual_error", data=df_class,
        jitter=True, size=5, edgecolor="black",
        palette=condition_color_palette, hue="condition",
        linewidth=0.5, alpha=0.6,
        ax=ax_boxplots, legend=False
    )

    medians = df_class.groupby("condition")["residual_error"].median()
    for i, cond in enumerate(boxplots_order):
        if cond in medians.index:
            y_med = medians[cond]
            ax_boxplots.text(
                i,
                y_med + 10,
                f"{y_med:.0f}k",
                ha="center",
                va="bottom",
                fontsize=9,
                color="black",
                bbox=dict(
                    boxstyle="round,pad=0.2",
                    facecolor="white",
                    alpha=0.8,
                    edgecolor="black",
                    linewidth=0.8,
                )
            )

    ax_boxplots.axhline(0, color="black", linestyle="--", linewidth=1.5, alpha=0.8)
    ax_boxplots.set_xlabel("")
    if idx == 0:
        ax_boxplots.set_ylabel("Residual error ($)", fontsize=12)
    else:
        ax_boxplots.set_ylabel("")
    ax_boxplots.set_title(complexity.capitalize(), fontsize=12)
    ax_boxplots.grid(True, axis="y", linestyle="--", alpha=0.5)
    ax_boxplots.tick_params(axis="x", labelsize=10)
    ax_boxplots.tick_params(axis="y", labelsize=12)

    ax_boxplots.set_xticks(ax_boxplots.get_xticks())
    ax_boxplots.set_xticklabels(
        [mapping_labels[l.get_text()] for l in ax_boxplots.get_xticklabels()]
    )

    ax_boxplots.yaxis.set_major_formatter(
        mtick.FuncFormatter(lambda x, pos: f"{int(x):,}k")
    )

plt.tight_layout()
plt.savefig(results_folder_path + "5_errors_by_complexity.pdf")
plt.show()
```

## 3. Statistical analysis
Where we evaluate the impact of task complexity and uncertainty visualization on estimation accuracy.    

### 3.1 Data preparation (long format df)

```python
cols = (
    id_cols +
    [col for col in df.columns if "house_value" in col] +
    [col for col in df.columns if "certainty_level" in col]
)
df_long = df.copy()[cols]

df_house_vals_list, df_house_certainty_list = [], []

for house_name in df_houses.name:
    house_name_lower = house_name.lower()

    # df with house values
    df_tmp = df_long.copy()
    df_tmp = (
        df_tmp[id_cols + [f"{house_name_lower}_house_value"]]
            .rename(columns={f"{house_name_lower}_house_value": "house_value"})
    )
    df_tmp["house_value"] = df_tmp["house_value"]
    df_tmp["house_name"] = house_name_lower
    df_tmp["complexity"] = df_houses[df_houses.name == house_name].complexity.values[0]
    df_tmp["true_value"] = df_houses[df_houses.name == house_name].true_value.values[0] * 100
    df_tmp["residual"] = df_tmp["house_value"] - df_tmp["true_value"]

    # df with certainty ratings
    df_tmp_cert = df_long.copy()
    df_tmp_cert = (
        df_tmp_cert[id_cols + [f"{house_name_lower}_certainty_level"]]
            .rename(columns={f"{house_name_lower}_certainty_level": "certainty_level"})
    )
    df_tmp_cert["house_name"] = house_name_lower

    # append dfs
    df_house_vals_list.append(df_tmp)
    df_house_certainty_list.append(df_tmp_cert)

# concat & merge
df_long = (
    pd.concat(df_house_certainty_list)
      .merge(pd.concat(df_house_vals_list), on=id_cols + ["house_name"])
)

df_long["abs_error"] = df_long["residual"].abs()

df_long.head()

# save data as input for future analysis
# df_long.to_csv('../data/interme
```

### 3.2 Normality assumption

```python
# Do data follow a normal distribution within groups of condition and difficulty level?
info_pvalues_norm = []

for c in df_long["complexity"].unique():
    print(f"{c.upper()}")
    
    for cond in df_long["condition"].unique():
        data = df_long.loc[
            (df_long["complexity"] == c) &
            (df_long["condition"] == cond),
            "residual"
        ]
        
        stat, p = shapiro(data)
        print(f"  {cond} → p(normality) = {p:.4f}")
        info_pvalues_norm.append([c, cond, p])
```

```python
print("Normality by complexity:")
for level in df_long["complexity"].unique():
    data = df_long.loc[df_long["complexity"] == level, "residual"]
    stat, p = shapiro(data)
    print(f"  {level}: p = {p:.4f}")

print("\nNormality by condition:")
for cond in df_long["condition"].unique():
    data = df_long.loc[df_long["condition"] == cond, "residual"]
    stat, p = shapiro(data)
    print(f"  {cond}: p = {p:.4f}")
```

```python
# df with p values
df_info_pvalues_norm = pd.DataFrame(
    info_pvalues_norm,
    columns=["complexity", "condition", "pvalue"]
)

diff_levels = sorted(df_long["complexity"].unique())
conditions = sorted(df_long["condition"].unique())

fig, axes = plt.subplots(
    len(diff_levels),
    len(conditions),
    figsize=(5 * len(conditions), 3 * len(diff_levels)),
    sharex=True,
    sharey=True
)

# garantisce array 2D anche con 1 sola riga/colonna
if len(diff_levels) == 1:
    axes = axes[np.newaxis, :]
if len(conditions) == 1:
    axes = axes[:, np.newaxis]

for i, d in enumerate(diff_levels):
    for j, c in enumerate(conditions):

        pval = df_info_pvalues_norm[
            (df_info_pvalues_norm.complexity == d) &
            (df_info_pvalues_norm.condition == c)
        ].pvalue.iloc[0].round(5)

        ax = axes[i, j]

        data = df_long.loc[
            (df_long["complexity"] == d) &
            (df_long["condition"] == c),
            "residual"
        ].dropna()

        ax.hist(data, bins=20, color="steelblue",
                edgecolor="black", alpha=0.7)
        ax.axvline(0, color="red", linestyle="--", linewidth=1)
        ax.set_title(f"{d} | {c} | {pval}", fontsize=10)
        ax.grid(linestyle="--", alpha=0.5)

plt.tight_layout()
plt.show()
```

### 3.3 Homogeneity of variances

```python
# Levene test by difficulty level
print("LEVENE – equal var across conditions:")
for d in df_long["complexity"].unique():
    groups = [
        df_long.loc[
            (df_long["condition"] == c) &
            (df_long["complexity"] == d),
            "residual"
        ]
        for c in df_long["condition"].unique()
    ]
    stat, p = levene(*groups)
    print(f"  {d.upper()}: p = {p:.4f}")

print("\n")
print("LEVENE – equal var across difficulty levels:")

# Levene test by condition
for c in df_long["condition"].unique():
    groups = [
        df_long.loc[
            (df_long["condition"] == c) &
            (df_long["complexity"] == d),
            "residual"
        ]
        for d in df_long["complexity"].unique()
    ]
    stat, p = levene(*groups)
    print(f"  {c.upper()}: p = {p:.4f}")
```

- For **conditions**: **p-values are all > 0.05**, meaning we cannot reject the null hypothesis of equal variances. Variability (spread of errors) across the three conditions appears similar within each difficulty level.
- For **complexity levels**: **p-values are all ≈ 0**, so the null hypothesis is rejected. It means variability differs significantly across difficulty levels within each visualization condition. As the task gets harder, the spread of residuals changes (often increases).

### 3.4 Kruskal test

```python
print("Kruskal across conditions:")
for cond in df_long["condition"].unique():
    groups = [
        df_long.loc[
            (df_long["complexity"] == c) &
            (df_long["condition"] == cond),
            "residual"
        ]
        for c in df_long["complexity"].unique()
    ]
    stat, p = kruskal(*groups)
    print(f"{cond}: H={stat:.3f}, p={p:.4f}")

print("\n")

print("Kruskal across difficulty levels:")
for complexity in df_long["complexity"].unique():
    groups = [
        df_long.loc[
            (df_long["complexity"] == complexity) &
            (df_long["condition"] == cond),
            "residual"
        ]
        for cond in df_long["condition"].unique()
    ]
    stat, p = kruskal(*groups)
    print(f"{complexity}: H={stat:.3f}, p={p:.4f}")
```

For easy and medium tasks, there’s a statistically significant difference between visualization types in terms of participants’ error (residuals).

→ The type of uncertainty display influences performance when the task isn’t too hard. For hard tasks, p = 0.0796 > 0.05 —that’s not significant, but borderline → we might just say that there's a weak trend (suggesting differences), but not strong enough to claim a clear effect.
