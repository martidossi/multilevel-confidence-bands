### 1.1 Visualizations

```python
df_certainty = df_ai[["condition", f"{sel_house.lower()}_certainty_level"]]

error_bars_certainty = list(df_certainty[df_certainty.condition=='Error bars'][f"{sel_house.lower()}_certainty_level"])
conf_bands_certainty = list(df_certainty[df_certainty.condition=='Confidence bands'][f"{sel_house.lower()}_certainty_level"])
no_ai_certainty = list(df_certainty[df_certainty.condition=='No AI'][f"{sel_house.lower()}_certainty_level"])
```

```python
categories = [1, 2, 3, 4, 5]
certainty_colors = ['#f2e9e4','#c9ada7','#9a8c98','#4a4e69','#22223b']
#certainty_colors = sns.color_palette("mako", 5)[::-1]

# Count occurrences for each complexity per question
certainty_tot = []
for q in [conf_bands_certainty, error_bars_certainty, no_ai_certainty]:
    c = Counter(q)
    certainty_tot.append([c.get(cat, 0) for cat in categories])

certainty_tot = np.array(certainty_tot)
certainty_tot_percent = certainty_tot / certainty_tot.sum(axis=1, keepdims=True) * 100

# Plot
fig, ax = plt.subplots(figsize=(6, 3))
left = np.zeros(len(certainty_tot_percent))
for _i in range(certainty_tot_percent.shape[1]):
    ax.barh(['Confidence bands', 'Error bars', 'No AI'], certainty_tot_percent[:, _i], left=left, color=certainty_colors[_i], label=categories[_i], height=0.6, alpha=0.8, edgecolor='black', linewidth=0.5)
    left += certainty_tot_percent[:, _i]

ax.set_xlabel('Percentage of responses')
ax.set_xticks(range(0, 101, 20), labels=[f'{el}%' for el in range(0, 101, 20)])
ax.set_title(f"{sel_house.capitalize()}\nCertainty raiting distribution", fontsize=10)
ax.legend(bbox_to_anchor=(1.05,1), loc='upper left')
ax.grid(True, axis='x', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.savefig(results_folder_path + f"/certainty_ratings/cert_rating_{sel_house.lower()}.pdf") #, dpi=600, bbox_inches="tight")
plt.show()
```

```python
easy_houses = ["Artemisstreet", "Aphroditestreet", "Apollostreet"][::-1]
medium_houses = ["Poseidonstreet", "Athenastreet", "Aresstreet"]
hard_houses = ["Hestiastreet", "Hephaestusstreet", "Zeusstreet"]

df_easy = df_ai[["condition"] + [f"{sel_house.lower()}_certainty_level" for sel_house in easy_houses]]
df_medium = df_ai[["condition"] + [f"{sel_house.lower()}_certainty_level" for sel_house in medium_houses]]
df_hard = df_ai[["condition"] + [f"{sel_house.lower()}_certainty_level" for sel_house in hard_houses]]

# --- Palettes and condition order ---
conditions = ['Confidence bands', 'Error bars', 'No AI']

# --- Collect data per condition (ABS counts and REL percentages) ---
if dropdown_complexity.value=='easy':
    df_cert_plot = df_easy
    house_names = easy_houses
if dropdown_complexity.value=='medium':
    df_cert_plot = df_medium
    house_names = medium_houses
if dropdown_complexity.value=='hard':
    df_cert_plot = df_hard
    house_names = hard_houses

list_error_bars_certainty = (
    df_cert_plot[df_cert_plot.condition=='Error bars']
    .drop('condition', axis=1)
    .transpose()
    .to_numpy()
    .tolist()
)
list_conf_bands_certainty = (
    df_cert_plot[df_cert_plot.condition=='Confidence bands']
    .drop('condition', axis=1)
    .transpose()
    .to_numpy()
    .tolist()
)
list_no_ai_certainty = (
    df_cert_plot[df_cert_plot.condition=='No AI']
    .drop('condition', axis=1)
    .transpose()
    .to_numpy()
    .tolist()
)

all_conditions = [
    list_conf_bands_certainty,
    list_error_bars_certainty,
    list_no_ai_certainty
]
```

```python
# 9 rows total: 3 questions per condition × 3 conditions
certainty_abs = []
for _condition in all_conditions:
    for _q in _condition:  # iterate over 3 rows per condition
        _c = Counter(_q)
        certainty_abs.append([_c.get(cat, 0) for cat in categories])

certainty_abs = np.array(certainty_abs)  # shape (9, n_categories)
certainty_rel = certainty_abs / certainty_abs.sum(axis=1, keepdims=True) * 100  # percentages

row_to_condition = [el for el in [[el]*3 for el in list(condition_palettes.keys())] for el in el]

# Only show the middle label of each triplet when interleaving, else we’ll use full labels or spaced labels later
yticklabels = house_names*3
# house_labels = [el for el in [[el]*3 for el in easy_houses] for el in el]
# yticklabels = [label if i in [1, 4, 7] else "" for i, label in enumerate(house_labels)]

# interleave the 3 groups: [0,3,6, 1,4,7, 2,5,8]
idx = np.array([2,1,0, 5,4,3, 8,7,6])
certainty_rel = certainty_rel[idx, :]
row_to_condition = [row_to_condition[i] for i in idx]
yticklabels = [yticklabels[i] for i in idx]

# --- Insert zero rows as spacers after every group of 3 rows ---
group_size = 3
spacer = np.zeros((certainty_rel.shape[1],))  # one zero row (per complexity stack)

certainty_rel_spaced = []
row_to_condition_spaced = []
labels_spaced = []
for _i in range(0, certainty_rel.shape[0], group_size):
    certainty_rel_spaced.extend(certainty_rel[_i:_i+group_size])
    row_to_condition_spaced.extend(row_to_condition[_i:_i+group_size])
    labels_spaced.extend(yticklabels[_i:_i+group_size])
    # Append a spacer row between groups (but not after the last one)
    if _i + group_size < certainty_rel.shape[0]:
        certainty_rel_spaced.append(spacer)
        row_to_condition_spaced.append("")   # mark as spacer (no color)
        labels_spaced.append("")             # blank label for spacer

certainty_rel_spaced = np.array(certainty_rel_spaced)  # shape (9 + n_groups-1, n_categories)
ypos = np.arange(certainty_rel_spaced.shape[0])

# --- Plot using the spaced arrays ---
fig, ax = plt.subplots(figsize=(4.5, 3.5))

_left = np.zeros(certainty_rel_spaced.shape[0])  # match spaced row count

for _i in range(certainty_rel_spaced.shape[1]):
    # Colors per row; 'none' for spacer rows so they don't draw
    colors_this_cat = [
        (condition_palettes[row_to_condition_spaced[_r]][_i]
         if row_to_condition_spaced[_r] in condition_palettes
         else 'none')
        for _r in range(certainty_rel_spaced.shape[0])
    ]

    ax.barh(
        ypos,
        certainty_rel_spaced[:, _i],
        color=colors_this_cat,
        left=_left,
        label=categories[_i],
        height=0.6,
        alpha=0.8,
        edgecolor='black',
        linewidth=0.5
    )
    _left += certainty_rel_spaced[:, _i]

# Grid, ticks, labels
ax.grid(True, axis='x', linestyle='--', alpha=0.5)

# Always use the spaced labels so tick count matches bars (11 positions: 9 rows + 2 spacers)
ax.set_yticks(ypos, labels_spaced)
ax.tick_params(axis='y', length=0)
plt.title(f"{dropdown_complexity.value.capitalize()}", fontsize=12)
ax.set_xlabel('Percentage of responses')
ax.set_xticks(range(0, 101, 20), labels=[f'{el}%' for el in range(0, 101, 20)])

patches = [
    mpatches.Patch(
        facecolor=c,
        edgecolor='black',   # black border
        linewidth=0.6,       # thin line
        label=l
    )
    for l, c in condition_color_palette.items()
]

ax.legend(
    handles=patches,
    loc='upper center',
    bbox_to_anchor=(0.5, -0.2),
    ncol=len(patches),
    frameon=False,           # no box around the whole legend
    handlelength=1,
    handleheight=1,
    columnspacing=1.5,
    borderpad=0.5
)

plt.tight_layout()
plt.savefig(results_folder_path + f"certainty_ratings_{dropdown_complexity.value}.pdf") #, dpi=600, bbox_inches="tight")
plt.show()
```

#### Differences among complexity levels and visualization conditions

```python
# Kruskal across complexity
groups = [
    df_long.dropna().loc[df_long['complexity'] == c, 'certainty_level']
    for c in ['easy', 'medium', 'hard']
]

stat, p = kruskal(*groups)
print(f"Kruskal (complexity effect): H={stat:.3f}, p={p:.4f}")

# Post-hoc Dunn
posthoc_diff = sp.posthoc_dunn(
    df_long,
    val_col='certainty_level',
    group_col='complexity',
    p_adjust='bonferroni'
)

print("\nDunn post-hoc by complexity:")
print(posthoc_diff.round(4))
```

```python
# Complexity
df_wide = df_long.pivot_table(index='participant_id', columns='complexity', values='certainty_level')
friedmanchisquare(df_wide['easy'], df_wide['medium'], df_wide['hard'])
```

```python
# Kruskal across condition
groups = [
    df_long.dropna().loc[df_long['condition'] == c, 'certainty_level']
    for c in ['No AI', 'Error bars', 'Confidence bands']
]

stat, p = kruskal(*groups)
print(f"Kruskal (condition effect): H={stat:.3f}, p={p:.4f}")

# Post-hoc Dunn
posthoc_diff = sp.posthoc_dunn(
    df_long,
    val_col='certainty_level',
    group_col='condition',
    p_adjust='bonferroni'
)

print("\nDunn post-hoc by condition:")
print(posthoc_diff.round(4))
```

#### Spearman correlation

```python
fig, axes = plt.subplots(1, 3, figsize=(13, 4), sharey=True)

for ax, diff in zip(axes, complexity_levels):
    df_sub = df_long[df_long["complexity"] == diff]

    for cond in ["No AI", "Error bars", "Confidence bands"]:
        color = condition_color_palette[cond]
        #for cond, color in condition_color_palette.items():
        df_cond = df_sub[df_sub["condition"] == cond]
        x = df_cond["certainty_level"].to_numpy()
        y = df_cond["abs_error"].to_numpy()

        # Fit
        slope, intercept, r, p, _ = linregress(x, y)
        x_grid = np.linspace(1, 5, 200)
        y_fit = intercept + slope * x_grid

        # --- Residual-based CI of the mean fit ---
        y_hat_obs = intercept + slope * x
        resid = y - y_hat_obs
        n = len(x)
        s2 = np.sum(resid**2) / (n - 2) # residual variance
        s = np.sqrt(s2) # residual std. error
        xbar = np.mean(x)
        Sxx = np.sum((x-xbar)**2)
        t_val = t.ppf(0.975, df=(n-2))

        se_mean = s * np.sqrt(1/n + (x_grid - xbar)**2 / Sxx)
        ci_low = y_fit - t_val * se_mean
        ci_high = y_fit + t_val * se_mean

        # Draw CI first so it sits beneath the line
        ax.fill_between(x_grid, ci_low, ci_high, color=color, alpha=0.25, zorder=1)

        # Scatter
        jitter = 0.12
        x_jitter = x + np.random.uniform(-jitter, jitter, size=len(x))
        # --- fixed per-condition offsets + tiny random jitter ---
        offset_by_cond = {
            "No AI": -0.15,
            "Error bars": 0.00,
            "Confidence bands": 0.15
        }
        inner_jitter = 0.06  # tiny randomness within each condition band
        # (inside your for cond, color in condition_color_palette.items(): loop)
        base = offset_by_cond.get(cond, 0.0)
        rng = np.random.default_rng(42)  # set once outside loop if you want reproducibility
        x_scatter = np.clip(
            x + base + rng.uniform(-inner_jitter, inner_jitter, size=len(x)),
            1-0.15, 5+0.15
        )
        ax.scatter(
            x_scatter, y, 
            s=30, alpha=0.6, color=color,
            edgecolors='black', linewidth=0.5,
            label=cond if ax is axes[0] else None, 
            zorder=2
        )

        # Regression line
        ax.plot(x_grid, y_fit, color=color, linewidth=2, zorder=3)

    ax.set_title(f"{diff.capitalize()}", fontsize=12)
    ax.set_xlabel("Certainty rating", fontsize=12)
    ax.set_xticks([1, 2, 3, 4, 5])
    ax.yaxis.set_major_formatter(
        mtick.FuncFormatter(lambda x, pos: f"{int(x):,}k")
    )
    ax.tick_params(axis='both', labelsize=12)

axes[0].set_ylabel("Absolute (residual) error ($)", fontsize=12)

#fig.legend(title="Condition", loc="center right", fontsize=12, title_fontsize=13)
axes[0].grid(True, axis='y', linestyle='--', alpha=0.5)
axes[1].grid(True, axis='y', linestyle='--', alpha=0.5)
axes[2].grid(True, axis='y', linestyle='--', alpha=0.5)

_handles, _labels = axes[0].get_legend_handles_labels()
import matplotlib.lines as mlines

_handles = [
    mlines.Line2D(
        [], [], color=condition_color_palette["No AI"], marker='o', linestyle='None', markersize=8,
        markeredgecolor='black', markeredgewidth=0.6),
    mlines.Line2D(
        [], [], color=condition_color_palette["Error bars"], marker='o', linestyle='None', markersize=8,
        markeredgecolor='black', markeredgewidth=0.6),
    mlines.Line2D(
        [], [], color=condition_color_palette["Confidence bands"], marker='o', linestyle='None', markersize=8,
        markeredgecolor='black', markeredgewidth=0.6)
]

fig.legend(
    _handles, ['No AI', 'Graded error bars', 'Multilevel confidence bands'],
    loc="lower center",
    ncol=3,
    bbox_to_anchor=(0.51, -0.1),
    fontsize=12,
    frameon=True
)

plt.tight_layout()
plt.savefig(results_folder_path + "6_confidence_and_error.pdf", bbox_inches='tight')
plt.show()

#Same with sns
#sns.lmplot(
#    data=df_long, x='certainty_level', y='abs_error',
#    hue='condition', col='difficulty_level',
#    scatter_kws={'alpha':0.5} #, line_kws={'color':'black'}
#)
#plt.show()
```

```python
# Spearman correlation
for c in df_long['complexity'].unique():
    subset = df_long[df_long['complexity'] == c].dropna()
    r, p = spearmanr(subset['certainty_level'], subset['abs_error'])
    print(f"{c}: rho = {r:.3f}, p = {p:.4f}")
```

```python
spearman_corr_list = []

for c in df_long['complexity'].unique():
    print(f"\n{c.upper()}")

    for cond in df_long['condition'].unique():
        subset = df_long[
            (df_long['complexity'] == c) &
            (df_long['condition'] == cond)
        ].dropna()

        x, y = subset['abs_error'], subset['certainty_level']

        def statistic(x_perm):  # permute only x
            return spearmanr(x_perm, y).statistic

        r, p = spearmanr(x, y)
        spearman_corr_list.append([r, cond, c])

        p_perm = permutation_test((x,), statistic, permutation_type='pairings').pvalue

        print(f"  {cond}: rho={r:.3f}, p={p:.4f}, p_perm={p_perm:.4f}")
```

```python
df_spearman_corr = pd.DataFrame(spearman_corr_list, columns=['value', 'condition', 'complexity'])
df_spearman_corr_mat = df_spearman_corr.pivot(index='condition', columns='complexity', values='value')
df_spearman_corr_mat[["easy", "medium", "hard"]]
```

```python
df_spearman_corr_mat = df_spearman_corr_mat[["easy", "medium", "hard"]]
plt.figure(figsize=(4.5, 3))
sns.heatmap(
    df_spearman_corr_mat,
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    center=0,
    vmin=-0.4,
    vmax=0.4,
    linewidths=0.5, 
    linecolor='white'  
)
plt.xlabel("")
plt.ylabel("")
plt.tight_layout()
plt.show()
```

## 2. AI-related questions

1. Which of the following statements, describes best how you made your decisions? `[Decision making]`
2. How much did you trust the AI's prediction? `[AI trust]`
3. Did the visualization help you confidently assess the AI's prediction uncertainties? `[AI support]`
4. Overall, how would you rate the usefulness of the AI prediction? `[AI usefulness]`

```python
decision_making_mapping = {
    'Mostly based on my own judgement ( ≤ 1 out of 9 cases)': 1,
    'Somewhat based on my own judgement (2-3 out of 9 cases)': 2,
    'Balanced mix of AI and my own judgement (4-5 out of 9 cases)': 3,
    "Somewhat based on the AI's prediction (6-7 out of 9 cases)": 4,
    "Mostly based on the AI's prediction ( ≥ 8 out of 9 cases)": 5,
}

decision_making_mapping_dict = {
    'a': 'Mostly based on my own judgement',
    'b': 'Somewhat based on my own judgement',
    'c': 'Balanced mix of AI and my own judgement',
    'd': "Somewhat based on the AI's prediction",
    'e': "Mostly based on the AI's prediction"
}

df_ai['decision_making_map'] = df_ai['decision_making'].map(decision_making_mapping)

def get_tot(df, col_name, categ=[1, 2, 3, 4, 5], perc=True):
    df_tmp = df[["condition", col_name]].dropna()
    error_bars_list = list(df_tmp[df_tmp.condition=='Error bars'][col_name])
    conf_bands_list = list(df_tmp[df_tmp.condition=='Confidence bands'][col_name])
    df_totals = []
    for val in [conf_bands_list, error_bars_list]:
        val_count = Counter(val)
        df_totals.append([val_count.get(cat, 0) for cat in categ])
    df_totals = np.array(df_totals)
    if perc:
        df_totals = df_totals / df_totals.sum(axis=1, keepdims=True) * 100
    return {
        'df_totals': df_totals,  
        'conf_bands_list': conf_bands_list,
        'error_bars_list': error_bars_list,
    }

def get_plot(
    df_tot,
    labels=[1, 2, 3, 4, 5], 
    colors=None, 
    ax=None,
    title=None
):
    df_totals = df_tot
    if colors is None:
        colors = ['#f2e9e4','#c9ada7','#9a8c98','#4a4e69','#22223b']

    if ax is None:
        fig, ax = plt.subplots(figsize=(6, 3))
    else:
        ax = ax
        fig = ax.figure

    _left = np.zeros(len(df_totals))
    for _i in range(df_totals.shape[1]):
        _y_pos = np.arange(2)
        bar_height = 0.4
        ax.barh(
            [0.25, 0.75],
            df_totals[:, _i],
            left=_left,
            color=colors[_i],
            label=labels[_i],
            height=0.3,
            alpha=0.8,
            edgecolor='black',
            linewidth=0.5
        )
        _left += df_totals[:, _i]
    ax.set_yticks([0.25, 0.75])
    ax.set_yticklabels(['Multilevel\nconfidence\nbands', 'Graded\nerror bars'])
    ax.set_ylim(0, 1)
    ax.set_xlabel('Percentage of responses', fontsize=18)
    ax.tick_params(axis='x', labelsize=18) 
    ax.tick_params(axis='y', labelsize=18) 
    ax.set_xticks(range(0, 101, 20), labels=[f'{el}%' for el in range(0, 101, 20)])
    if title:
        ax.set_title(title, fontsize=24)
    else:
        ax.set_title(col_name, fontsize=10)        
    ax.legend(bbox_to_anchor=(1.05,1), loc='upper left', fontsize=18)
    ax.grid(True, axis='x', linestyle='--', alpha=0.5)   
    fig.text(
        0.5, -0.05,
        "a: Mostly based on my own judgement | "
        "b: Somewhat based on my own judgement\n"
        "c: Balanced mix | "
        "d: Somewhat based on AI | "
        "e: Mostly based on AI",
        ha='center', va='center', fontsize=20
    )
    fig.tight_layout()
```

```python
df_house_properties = df_houses.copy()
df_house_properties['name'] = df_house_properties['name'].str.lower()

df_house_melted = (
    df[['participant_id'] + [col for col in df.columns if 'house_value' in col]]
        .drop(['training_house_value'], axis=1)
        .melt(
            id_vars='participant_id',
            value_vars=[col for col in df.columns if 'house_value' in col if col!="training_house_value"],
            var_name='house_name',
            value_name='house_value'
        )
)
df_house_melted['house_name'] = df_house_melted['house_name'].str.replace('_house_value', '', regex=False)
df_house_melted['house_value'] = df_house_melted['house_value']/100
df_house_melted = df_house_melted.rename(columns={'house_value': 'house_pred'})
df_house_melted = df_house_melted.merge(df_house_properties[['name', 'true_value']], left_on='house_name', right_on='name', how='left').drop(['name'], axis=1)

df_house_melted['residual_error'] = df_house_melted["true_value"] - df_house_melted["house_pred"]
df_house_melted['abs_error'] = np.abs(df_house_melted['residual_error'])
df_house_melted[df_house_melted.participant_id=='participant_1']
```
